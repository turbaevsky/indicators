```r
source('shiny/data.r')
source('functions.r')
data <- data.frame(stringsAsFactors=FALSE)
#place
#relation
#relType
uByCentre()
PC <- intersect(unitsByCentre$uList[[3]], activeStation(201712))
dates <- unlist(unique(subset(idx, LocId %in% PC & PeriodEndYrMn >= 200200, PeriodEndYrMn)))
idx <- subset(idx, LocId %in% PC & NonQualCode == ' ' & PeriodEndYrMn >= 200200 & IndexId == 4, c(LocId,PeriodEndYrMn,IndexValue))
for (u in PC){
    util <- unlist(subset(place,LocId==tail(unlist(subset(relation,LocId==u & RelationId==3,ParentLocId)),1),AbbrevLocName))
    row <- data.frame(util,nameByID(u))
    for (d in dates){
        index <- unlist(subset(idx,LocId == u & PeriodEndYrMn == d, IndexValue))
        if (length(index) && is.numeric(index)) Idx <- index
        else Idx <- 0
        row <- data.frame(row,Idx)
    }
    data <- rbind(data,row)
}
colnames(data) <- c('Utility','Unit',dates)
rownames(data) <- NULL
write.csv(data,'PCIndexes.csv')
```
